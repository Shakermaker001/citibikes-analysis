```python

# Get the current directory
current_directory = os.getcwd()

# Get a list of all CSV files in the current directory
csv_files = [file for file in os.listdir(current_directory) if file.endswith('.csv')]

# Create an empty dictionary to store the dataframes
dataframes = {}

# Iterate over the CSV files and create dataframes
for file in csv_files:
    # Get the file path
    file_path = os.path.join(current_directory, file)
    
    # Read the CSV file and create a dataframe
    df = pd.read_csv(file_path)
    
    # Add the dataframe to the dictionary with the file name as the key
    dataframes[file] = df

# Print the dataframes
for file, df in dataframes.items():
    print(f"Dataframe from {file}:")
    print(df)
    print()
```

```python
filtered_rides = {}

for month, rides_df in dataframes.items():
    # Check if 'rideable_type' is present in the DataFrame
    if 'rideable_type' in rides_df.columns:
        # Define your conditions for each month
        conditions = (rides_df['rideable_type'] == 'classic_bike') & (rides_df['member_casual'] == 'casual')     
        
        # Apply conditions using .loc and store the result in the dictionary
        filtered_rides[month] = rides_df.loc[conditions].drop(['rideable_type', 'member_casual'], axis=1).dropna()

january_rides_df = filtered_rides.get('JC-202301-citibike-tripdata.csv', pd.DataFrame())
february_rides_df = filtered_rides.get('JC-202302-citibike-tripdata.csv', pd.DataFrame())
march_rides_df = filtered_rides.get('JC-202303-citibike-tripdata.csv', pd.DataFrame())
april_rides_df = filtered_rides.get('JC-202304-citibike-tripdata.csv', pd.DataFrame())
may_rides_df = filtered_rides.get('JC-202305-citibike-tripdata.csv', pd.DataFrame())
june_rides_df = filtered_rides.get('JC-202306-citibike-tripdata.csv', pd.DataFrame())
july_rides_df = filtered_rides.get('JC-202307-citibike-tripdata.csv', pd.DataFrame())
august_rides_df = filtered_rides.get('JC-202308-citibike-tripdata.csv', pd.DataFrame())
september_rides_df = filtered_rides.get('JC-202309-citibike-tripdata.csv', pd.DataFrame())
october_rides_df = filtered_rides.get('JC-202310-citibike-tripdata.csv', pd.DataFrame())
november_rides_df = filtered_rides.get('JC-202311-citibike-tripdata.csv', pd.DataFrame())
december_rides_df = filtered_rides.get('JC-202312-citibike-tripdata.csv', pd.DataFrame())

december_rides_df.head()
```

```python
# Merge the dataframes for summer months
summer_months = [june_rides_df, july_rides_df, august_rides_df]
summer_df = pd.concat(summer_months)

# Merge the dataframes for winter months
winter_months = [december_rides_df, january_rides_df, february_rides_df]
winter_df = pd.concat(winter_months)

# Write the summer dataframe to a .csv file
summer_df.to_csv('summer_rides2023.csv', index=False)

# Write the winter dataframe to a .csv file
winter_df.to_csv('winter_rides2023.csv', index=False)
```

```python
winter_time_df = pd.read_csv('winter_rides.csv')
summer_time_df = pd.read_csv('summer_rides.csv')
# Convert the 'started_at' and 'ended_at' columns to datetime
winter_time_df['started_at'] = pd.to_datetime(winter_time_df['started_at'])
winter_time_df['ended_at'] = pd.to_datetime(winter_time_df['ended_at'])

summer_time_df['started_at'] = pd.to_datetime(summer_time_df['started_at'])
summer_time_df['ended_at'] = pd.to_datetime(summer_time_df['ended_at'])

# Calculate the ride duration by subtracting 'started_at' from 'ended_at'
winter_time_df['ride_duration'] = (winter_time_df['ended_at'] - winter_time_df['started_at']).dt.total_seconds()
summer_time_df['ride_duration'] = (summer_time_df['ended_at'] - summer_time_df['started_at']).dt.total_seconds()

# Convert ride duration to hours, minutes, and seconds
winter_time_df['ride_duration'] = pd.to_datetime(winter_time_df['ride_duration'], unit='s').dt.strftime('%H:%M:%S')
summer_time_df['ride_duration'] = pd.to_datetime(summer_time_df['ride_duration'], unit='s').dt.strftime('%H:%M:%S')

# Dropping unnecessary columns
winter_time_df = winter_time_df.drop(['start_station_name', 'end_station_name', 'start_lat', 'start_lng', 'end_lat', 'end_lng'], axis=1)
summer_time_df = summer_time_df.drop(['start_station_name', 'end_station_name', 'start_lat', 'start_lng', 'end_lat', 'end_lng'], axis=1)

# Reorder columns
winter_time_df = winter_time_df.reindex(columns=['ride_id','started_at', 'ended_at','ride_duration','start_station_id','end_station_id']).dropna()
summer_time_df = summer_time_df.reindex(columns=['ride_id','started_at', 'ended_at','ride_duration','start_station_id','end_station_id']).dropna()

winter_time_df.to_csv('winter_time2023.csv', index=False)
summer_time_df.to_csv('summer_time2023.csv', index=False)
```

```python
winter_df = pd.read_csv('winter_rides.csv')
summer_df = pd.read_csv('summer_rides.csv')

# Create a dataframe based on the value counts of the start station_name
winter_start = pd.DataFrame({'start_station_name': winter_df['start_station_name'].value_counts().index,
                             'start_count': winter_df['start_station_name'].value_counts().values})

# Create a dataframe based on the value counts of the end station_name
winter_end = pd.DataFrame({'end_station_name': winter_df['end_station_name'].value_counts().index,
                           'end_count': winter_df['end_station_name'].value_counts().values})

# Merge the start and end station counts dataframes based on the station_name column
winter_counts = pd.merge(winter_start, winter_end, left_on='start_station_name', right_on='end_station_name')

# Drop the duplicate station_name column
winter_counts.drop('end_station_name', axis=1, inplace=True)

# Merge the latitude, longitude, and station_id columns from the original dataframe
winter_counts = pd.merge(winter_counts, winter_df[['start_station_name', 'start_lat', 'start_lng', 'start_station_id']], on='start_station_name').drop_duplicates(['start_station_name']).dropna()

# Rename the columns
winter_counts.rename(columns={'start_station_id': 'station_id',
                              'start_station_name': 'station_name',
                              'start_count': 'start_count',
                              'end_count': 'end_count',
                              'start_lat': 'latitude',
                              'start_lng': 'longitude',
                              }, inplace=True)

# Print the resulting dataframe
winter_counts.to_csv('winter_counts2023.csv', index=False)
```

```python
summer_df = pd.read_csv('summer_rides.csv')
# Create a dataframe based on the value counts of the start station_name

summer_start = pd.DataFrame({'start_station_name': summer_df['start_station_name'].value_counts().index,
                                     'start_count': summer_df['start_station_name'].value_counts().values})

# Create a dataframe based on the value counts of the end station_name
summer_end = pd.DataFrame({'end_station_name': summer_df['end_station_name'].value_counts().index,
                                   'end_count': summer_df['end_station_name'].value_counts().values})

# Merge the start and end station counts dataframes based on the station_name column
summer_counts = pd.merge(summer_start, summer_end, left_on='start_station_name', right_on='end_station_name')

# Drop end_station_name column
summer_counts.drop('end_station_name', axis=1, inplace=True)

# Merge the latitude and longitude columns from the original dataframe

summer_counts = pd.merge(summer_counts, summer_df[['start_station_name', 'start_lat', 'start_lng','start_station_id']], on='start_station_name').drop_duplicates(['start_station_name']).dropna()


# Rename the columns
summer_counts.rename(columns={'start_station_id': 'station_id',
                              'start_station_name': 'station_name',
                              'start_count': 'start_count',
                              'end_count': 'end_count',
                              'start_lat': 'latitude',
                              'start_lng': 'longitude',
                              }, inplace=True)

# Print the resulting dataframe
summer_counts.to_csv('summer_counts2023.csv', index=False)
```
