```python

# Get the current directory
current_directory = os.getcwd()

# Get a list of all CSV files in the current directory
csv_files = [file for file in os.listdir(current_directory) if file.endswith('.csv')]

# Create an empty dictionary to store the dataframes
dataframes = {}

# Iterate over the CSV files and create dataframes
for file in csv_files:
    # Get the file path
    file_path = os.path.join(current_directory, file)
    
    # Read the CSV file and create a dataframe
    df = pd.read_csv(file_path)
    
    # Add the dataframe to the dictionary with the file name as the key
    dataframes[file] = df

# Print the dataframes
for file, df in dataframes.items():
    print(f"Dataframe from {file}:")
    print(df)
    print()
```

```python
filtered_rides = {}

for month, rides_df in dataframes.items():
    # Check if 'rideable_type' is present in the DataFrame
    if 'usertype' in rides_df.columns:
        # Define your conditions for each month
        conditions = (rides_df['usertype'] == 'Customer')
        
        # Apply conditions using .loc and store the result in the dictionary
        filtered_rides[month] = rides_df.loc[conditions].drop(['usertype', 'gender','bikeid','birth year'], axis=1).dropna()

january_rides_df = filtered_rides.get('201801-citibike-tripdata.csv', pd.DataFrame())
february_rides_df = filtered_rides.get('201802-citibike-tripdata.csv', pd.DataFrame())
march_rides_df = filtered_rides.get('201803-citibike-tripdata.csv', pd.DataFrame())
april_rides_df = filtered_rides.get('201804-citibike-tripdata.csv', pd.DataFrame())
may_rides_df = filtered_rides.get('201805-citibike-tripdata.csv', pd.DataFrame())
june_rides_df = filtered_rides.get('201806-citibike-tripdata.csv', pd.DataFrame())
july_rides_df = filtered_rides.get('201807-citibike-tripdata.csv', pd.DataFrame())
august_rides_df = filtered_rides.get('201808-citibike-tripdata.csv', pd.DataFrame())
september_rides_df = filtered_rides.get('201809-citibike-tripdata.csv', pd.DataFrame())
october_rides_df = filtered_rides.get('201810-citibike-tripdata.csv', pd.DataFrame())
november_rides_df = filtered_rides.get('201811-citibike-tripdata.csv', pd.DataFrame())
december_rides_df = filtered_rides.get('201812-citibike-tripdata.csv', pd.DataFrame())

december_rides_df.head()
```

```python
# Merge the dataframes for summer months
summer_months = [june_rides_df, july_rides_df, august_rides_df]
summer_df = pd.concat(summer_months)

# Merge the dataframes for winter months
winter_months = [december_rides_df, january_rides_df, february_rides_df]
winter_df = pd.concat(winter_months)

# Write the summer dataframe to a .csv file
summer_df.to_csv('summer_rides2018.csv', index=False)

# Write the winter dataframe to a .csv file
winter_df.to_csv('winter_rides2018.csv', index=False)
```

```python
winter_time_df = pd.read_csv('winter_rides2018.csv')
summer_time_df = pd.read_csv('summer_rides2018.csv')
# Convert the 'starttime' and 'stoptime' columns to datetime
winter_time_df['starttime'] = pd.to_datetime(winter_time_df['starttime'])
winter_time_df['stoptime'] = pd.to_datetime(winter_time_df['stoptime'])

summer_time_df['starttime'] = pd.to_datetime(summer_time_df['starttime'])
summer_time_df['stoptime'] = pd.to_datetime(summer_time_df['stoptime'])

# Calculate the ride duration by subtracting 'starttime' from 'stoptime'
winter_time_df['ride_duration'] = (winter_time_df['stoptime'] - winter_time_df['starttime']).dt.total_seconds()
summer_time_df['ride_duration'] = (summer_time_df['stoptime'] - summer_time_df['starttime']).dt.total_seconds()

# Convert ride duration to hours, minutes, and seconds
winter_time_df['ride_duration'] = pd.to_datetime(winter_time_df['ride_duration'], unit='s').dt.strftime('%H:%M:%S')
summer_time_df['ride_duration'] = pd.to_datetime(summer_time_df['ride_duration'], unit='s').dt.strftime('%H:%M:%S')

# Reorder columns
winter_time_df = winter_time_df.reindex(columns=['starttime', 'stoptime','ride_duration','start station id','end station id']).dropna()
summer_time_df = summer_time_df.reindex(columns=['starttime', 'stoptime','ride_duration','start station id','end station id']).dropna()

winter_time_df.to_csv('winter_time2018.csv', index=False)
summer_time_df.to_csv('summer_time2018.csv', index=False)
```

```python
winter_df = pd.read_csv('winter_rides2018.csv')

# Create a dataframe based on the value counts of the start station_name
winter_start = pd.DataFrame({'start station name': winter_df['start station name'].value_counts().index,
                             'start_count': winter_df['start station name'].value_counts().values})

# Create a dataframe based on the value counts of the end station_name
winter_end = pd.DataFrame({'end station name': winter_df['end station name'].value_counts().index,
                           'end_count': winter_df['end station name'].value_counts().values})

# Merge the start and end station counts dataframes based on the station_name column
winter_counts = pd.merge(winter_start, winter_end, left_on='start station name', right_on='end station name')

# Drop the duplicate station_name column
winter_counts.drop('end station name', axis=1, inplace=True)

# Merge the latitude, longitude, and station_id columns from the original dataframe
winter_counts = pd.merge(winter_counts, winter_df[['start station name', 'start station latitude', 'start station longitude', 'start station id']],
                          on='start station name').drop_duplicates(['start station name']).dropna()

# Rename the columns
winter_counts.rename(columns={'start station id': 'station_id',
                              'start station name': 'station_name',
                              'start_count': 'start_count',
                              'end_count': 'end_count',
                              'start station latitude': 'latitude',
                              'start station longitude': 'longitude',
                              }, inplace=True)

# Print the resulting dataframe
winter_counts.to_csv('winter_counts2018.csv', index=False)
```
